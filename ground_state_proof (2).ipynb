```python
from sympy import *
from sympy.physics.quantum import TensorProduct
init_printing(pretty_print=True)

h,k = symbols('h, k')
id= Matrix([[1,0],[0,1]])
Z= Matrix([[1,0],[0,-1]])
X= Matrix([[0,1],[1,0]])
Z0= TensorProduct(Z, id, id)
Z1= TensorProduct(id, Z, id)
Z2= TensorProduct(id, id ,Z)
I = TensorProduct(id, id, id)
X0X1 = TensorProduct(X, X, id)
X1X2 = TensorProduct(id, X, X)
X0X2 = TensorProduct(X, id, X)


H_tot= h*(Z0+Z1+Z2)+k*(X0X1+X1X2+X0X2)+(2*sqrt(h**2 + h*k + k**2)+h-k)*I;
```

```python
H_tot
```

```python
H_tot.eigenvals()
```

the eigen vals output a:b means that a is the eigenvalue and b is multiplicity 

```python
H_tot.subs({h:1, k:1}).eigenvals()
```

the nullspace fucntion can output the eigenvector corresponding to 0 only


```python
H_tot.nullspace()
```

```python
H_tot.subs({h:1, k:1}).nullspace()
```
